```python
country_uses_python = source_data.groupby(["Country"])["LanguageWorkedWith"].apply(lambda x : x.str.contains("Python").sum())
```

```python
# Percentage of people use python in each country

country_respondents = source_data["Country"].value_counts()
```

```python
import pandas as pd
```

```python

python_df = pd.concat([country_respondents, country_uses_python], axis = 'columns', sort=False)
```

```python
python_df.rename(columns={"Country": "NumRespondents", "LanguageWorkedWith" : "NumKnowsPython"}, inplace=True)
```

```python
python_df.index.rename("Country", inplace=True)
```

```python
python_df["PctKnowsPythnon"] = python_df["NumKnowsPython"] / python_df["NumRespondents"]
```

```python
python_df.sort_values(by="PctKnowsPythnon", ascending=False).loc["Japan"]
```

```python
na_vals = ["NA", "Missing"]
df = pandas.read_csv("developer_survey_2019/survey_results_public.csv", na_values=na_vals)
```

```python
df["YearsCode"].count_values()
```

```python
df["YearsCode"].replace("Less than 1 year", 0, inplace=True)
df["YearsCode"].replace("More than 50 years", 51, inplace=True)
```

```python
df["YearsCode"] = df["YearsCode"].astype(float)
```
